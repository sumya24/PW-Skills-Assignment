## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer:

Line Plot (sns.lineplot)

Use: Shows the relationship between two continuous variables. Often used to visualize trends over time.

Bar Plot (sns.barplot)

Use: Shows the average (mean) value of a variable for each category of another variable.

Box Plot (sns.boxplot)

Use: Displays the distribution of data based on five summary statistics: min, Q1, median, Q3, and max. Great for detecting outliers.

Histogram (sns.histplot)

Use: Shows the distribution of a single variable. Helps understand the frequency of data ranges.

Heatmap (sns.heatmap)

Use: Displays data in a matrix format using color intensity. Useful for correlation matrices or pivot tables.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
fmri = sns.load_dataset("fmri")

# Line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Box plot 1 - Age vs Pclass
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Age Distribution by Passenger Class")

# Box plot 2 - Fare vs Pclass
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Fare Distribution by Passenger Class")

plt.tight_layout()
plt.show()
```

##  Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Load Diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Histogram of price with hue as cut
sns.histplot(data=diamonds, x='price', hue='cut', kde=True, palette='Set2', element='step')
plt.title("Price Distribution by Cut Type")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
# Load Iris dataset
iris = sns.load_dataset("iris")

# Pair plot
sns.pairplot(iris, hue="species", palette="Set1")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

##  Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Load Flights dataset
flights = sns.load_dataset("flights")

# Pivot the data for heatmap
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Monthly Passenger Count (1949 - 1960)")
plt.show()
```
