## Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Answer: Estimation statistics involves using sample data to estimate population parameters. It helps us infer or predict characteristics of a population.

Point Estimate: A single value used to estimate a population parameter (e.g., sample mean to estimate population mean).

Interval Estimate: A range of values within which a population parameter is expected to lie, typically given with a confidence level (e.g., 95% confidence interval).



## Q2: Write a Python function to estimate the population mean using a sample mean and standard deviation.

```python
import scipy.stats as stats
import numpy as np

def estimate_population_mean(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std = np.std(sample, ddof=1)
    se = std / np.sqrt(n)
    margin = stats.t.ppf((1 + confidence) / 2, df=n-1) * se
    return (mean - margin, mean + margin)

# Example
sample = [480, 500, 520, 510, 495, 505]
print("95% Confidence Interval:", estimate_population_mean(sample))
```

## Q5: Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

```python
from scipy.stats import ttest_ind

group1 = [80, 85, 78, 90, 88]
group2 = [75, 70, 74, 72, 73]

t_stat, p_value = ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

## Q8: Define p-value and explain its significance in hypothesis testing.

Answer: P-value is the probability of obtaining results at least as extreme as the observed ones, assuming the null hypothesis is true.

Significance:

Small p-value (< α): Reject H₀.

Large p-value (≥ α): Fail to reject H₀.



## Q9: Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

df = 10
x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df)

plt.plot(x, y)
plt.title("Student's t-Distribution (df=10)")
plt.xlabel('t')
plt.ylabel('Probability Density')
plt.grid()
plt.show()
```

## Q13: Estimate the population mean revenue with a 95% confidence interval.

```python
import scipy.stats as stats
import numpy as np

mean = 500
std = 50
n = 50
confidence = 0.95

se = std / np.sqrt(n)
margin = stats.t.ppf((1 + confidence) / 2, df=n-1) * se
lower = mean - margin
upper = mean + margin

print(f"95% Confidence Interval: (${lower:.2f}, ${upper:.2f})")
```

## Q14: Test the hypothesis that the new drug decreases blood pressure by 10 mmHg.

```python
import scipy.stats as stats
import numpy as np

sample_mean = 8
population_mean = 10
std_dev = 3
n = 100

se = std_dev / np.sqrt(n)
t_stat = (sample_mean - population_mean) / se
p_value = stats.t.sf(np.abs(t_stat), df=n-1) * 2

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

## Q15: Test if true mean weight is less than 5 pounds.

```python
mean = 4.8
pop_mean = 5
std = 0.5
n = 25

se = std / np.sqrt(n)
t_stat = (mean - pop_mean) / se
p_value = stats.t.cdf(t_stat, df=n-1)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

## Q16: Test if two groups' population means are equal.

```python
mean1, std1, n1 = 80, 10, 30
mean2, std2, n2 = 75, 8, 40

se = np.sqrt((std1**2 / n1) + (std2**2 / n2))
t_stat = (mean1 - mean2) / se
df = min(n1, n2) - 1
p_value = stats.t.sf(np.abs(t_stat), df) * 2

print("T-statistic:", t_stat)
print("P-value:", p_value)
```
